# EXCEPTION HANDLING
## LAB REPORT TASK
You are developing a Student Result Processing System for a small institute.
The program takes student marks as input and calculates results. Since users may enter incorrect data, the program must handle errors properly using Python exception handling.
## Task 1: Handle Invalid User Input
Problem
The program asks the user to enter numeric values for total marks and obtained marks. Users may enter invalid input such as letters or symbols.
The program must not crash and should display a clear error message.
Use the Following Concepts
try
except
ValueError

```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))
except ValueError:
    print("Error: Please enter numeric value only")
```

---

## Task 2: Handle Division by Zero
Problem
If total marks are entered as zero, calculating the percentage will result in a runtime error.
Modify the program to safely handle this case and notify the user.
Use the Following Concepts
    ZeroDivisionError
	Multiple except blocks

```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))
    percentage = (obtained / total) * 100
except ZeroDivisionError:
    print("Error: Total marks cannot be zero.")
except ValueError:
    print("Error: Invalid input.")
```

---

## Task 3: Validate Logical Conditions Using raise
Problem
Obtained marks must not be negative and must not exceed total marks.
If the condition is violated, the program should manually raise an exception with a meaningful message.
Use the Following Concepts
raise
Exception

```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))

    if obtained < 0 or obtained > total:
        raise Exception("Obtained marks must be between 0 and total marks.")
except Exception as e:
    print("Error:", e)
```

---

### Task 4: Execute Code Only When No Error Occurs
Problem
Display the calculated percentage and pass or fail status only if no exception occurs during input and validation.
Use the Following Concepts
try
except
else


```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))

    if obtained < 0 or obtained > total:
        raise Exception("Invalid marks entered.")

    percentage = (obtained / total) * 100
except Exception as e:
    print("Error:", e)
else:
    print("Percentage:", percentage)
    print("Result:", "Pass" if percentage >= 50 else "Fail")
```

### Task 5: Guarantee Program Completion Using finally
Problem
No matter what happens during execution, the program must display:
Result processing completed.
This ensures proper program termination behavior.
Use the Following Concepts
finally

```python
try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Invalid input. Please enter a number.")

finally:
    print("Result processing completed.")
```
