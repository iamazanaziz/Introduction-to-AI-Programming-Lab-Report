## Lab Report Task:
You are developing a simple energy consumption logging system for a residential building.
Daily electricity usage is stored in a CSV file. so it can be easily read, updated, and organized.


## Task 1: Create and Write Energy Data to a CSV File
Create a file named energy_data.csv.
Write the header and daily energy consumption records into the file using the CSV format:
	Date
	Energy consumed in kWh
Each record must be written on a new line following the header.


```python
import csv

with open("energy_data.csv", mode="w", newline="") as file:
    writer = csv.writer(file)

    # Writing header
    writer.writerow(["Date", "Energy Consumed (kWh)"])

    # Writing daily energy records
    writer.writerow(["2026-01-01", 8.0])
    writer.writerow(["2026-01-02", 12.8])
    writer.writerow(["2026-01-03", 15.9])

print("Energy data written successfully.")
```

## Task 2: Read and Display CSV File Data
Read the contents of energy_data.csv and display all energy consumption records in a clear and readable format.
The data must be read from the file, not hardcoded.


```python
import csv

with open("energy_data.csv", mode="r") as file:
    reader = csv.reader(file)

    print("\nEnergy Consumption Records:")
    for row in reader:
        print(row[0], "→", row[1])
```
