## Extra Question (Using both FOR and WHILE loops)

An engineer monitors the charging process of multiple batteries.
Each battery has a target charge level.

Use a for loop to go through each battery.

For each battery, use a while loop to simulate the charging process until the battery reaches its target level.

```python
# Target charge levels of batteries (in %)
battery_targets = [2, 5, 4]

# For loop to check each battery
for target in battery_targets:
    print("Charging new battery")

    charge = 0
    # While loop to simulate charging
    while charge < target:
        charge = charge + 1
        print("Battery charge:", charge, "%")

    print("Battery fully charged\n")
```
